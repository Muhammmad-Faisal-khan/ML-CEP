# split into X(feature) and y(Lable)

```python
X=dataset.drop(["status" , "name"], axis=1)
X.shape
```

```python
y=dataset["status"]
y.shape
```

# Split Data into Train and Test

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

# Balance The unbalance Data

```python
from sklearn.utils.class_weight import compute_sample_weight
w_train = compute_sample_weight('balanced', y_train)
print(w_train, y_train)
```

# Provide Scalling to Train and Test Data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
sc.fit(X_train)
X_train= sc.transform(X_train)
X_test= sc.transform(X_test)
```

# Apply the Logistic Model

```python
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
log.fit(X_train, y_train, sample_weight=w_train)
```

# Balance and Unbalance Score of Logistic Regression

```python
log.score(X_train, y_train)
```

```python
log.score(X_train, y_train, sample_weight=w_train)
```

```python
y_pred = log.predict(X_test)
```

# Finding F1_Score (Logistic Regression) 

```python
from sklearn.metrics import f1_score, accuracy_score
f1_score(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```

# Train and Test score for Logistic Regression

```python
LogisticScoreTrain = log.score(X_train, y_train)
LogisticScoreTest = log.score(X_test, y_test)
print("LogisticScoreTrain = ",LogisticScoreTrain)
print("LogisticScoreTest = ",LogisticScoreTest)
```

# Apply Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
DTree = DecisionTreeClassifier(criterion= 'gini' , max_depth=5)
```

```python
DTree.fit(X_train, y_train)
```

```python
DTree.fit(X_train, y_train, sample_weight=w_train)
```

# Score on Balance and Imbalance Data

```python
DTree.score(X_train, y_train, sample_weight=w_train)
```

```python
DTree.score(X_test, y_test)
```

```python
DTree.score(X_train, y_train)
```

```python
Prediction = DTree.predict(X_train)
```

# F1_Score for Decision Tree

```python
from sklearn.metrics import f1_score
f1 = f1_score(Prediction, y_train)
f1
```

# Apply SVC Model

```python
from sklearn.svm import SVC
SVC_rbf = SVC(kernel = "rbf") #Radial basis function
```

```python
SVC_rbf.fit(X_train, y_train)
```

# Score by Radial Basis Function

```python
SVC_rbf.score(X_test, y_test)
```

```python
SVC_poly = SVC(kernel='poly', degree=2)
SVC_poly.fit(X_train, y_train)
```

```python
SvcPolytest = SVC_poly.score(X_test, y_test)
SvcPolytrain = SVC_poly.score(X_train, y_train)
print(SvcPolytrain)
print(SvcPolytest)
```
